```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
result_completo = pd.read_excel('resultados/resultado_consolidado.xlsx')
result_completo.head()
```

```python
df_media = result_completo.groupby(['banco_de_dados','consulta'])['tempo'].mean().reset_index()
# Get colors from the 'Set1' palette
colors = sns.color_palette('Set1', len(df_media['banco_de_dados'].unique()))

# Create a dictionary to map categories to colors
color_dict = dict(zip(df_media['banco_de_dados'].unique(), colors))
```

```python
def plota_graficos_consulta(num_query: int):

    df_query = result_completo[result_completo['consulta'] == f'query{num_query}']

    # Plota o gráfico de linhas que exibe o tempo de cada execução da query
    sns.lineplot(data=df_query, x='num_execucao', y='tempo', marker='o', hue='banco_de_dados', palette=color_dict)
    plt.xlabel('Número da execução')
    plt.ylabel('Tempo (segundos)')
    plt.title(f'Query {num_query} - Tempo de execução')
    plt.xticks(df_query['num_execucao'].unique())
    plt.show()

    df_media_query = df_media[df_media['consulta'] == f'query{num_query}']
    print('=== Tempo médio ===')
    print(df_media_query)

    # Plota o gráfico de barras que compara a média do tempo da consulta em cada banco de dados
    sns.barplot(data=df_media_query, x='banco_de_dados', y='tempo', palette=color_dict, hue='banco_de_dados', width=0.5)
    plt.xlabel('Banco de Dados')
    plt.ylabel('Tempo médio')
    plt.title(f'Query {num_query} - Comparação do tempo médio')
    plt.show()
```

```python
def plota_grafico_consolidado():
    df_media_bd = result_completo.groupby(['banco_de_dados'])['tempo'].mean().reset_index()
    sns.barplot(data=df_media_bd, x='banco_de_dados', y='tempo', palette=color_dict, hue='banco_de_dados', width=0.5)
    plt.xlabel('Banco de Dados')
    plt.ylabel('Tempo médio (segundos)')
    plt.title(f'Tempo médio por banco de dados')
    plt.show()
```

# Resultado Consolidado

```python
plota_grafico_consolidado()
```
